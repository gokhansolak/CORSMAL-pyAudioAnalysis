```python
import pyAudioAnalysis
from pyAudioAnalysis import audioTrainTest as aT
import plotly.subplots
dataset_path="../datasets"
```

```python
def train_fold(fold_no):

    model_name = "fillings-audio-3fold"

    data_folders = [dataset_path+"/"+model_name+"/train"+str(fold_no)+"/fi"+str(i) for i in range(4)]

    aT.extract_features_and_train(data_folders, 1.0, 1.0, aT.shortTermWindow, aT.shortTermStep, "svm", model_name+str(fold_no), False)
```

```python
def test_fold(fold_no):
    test_folders = [dataset_path+"/"+model_name+"/test"+str(fold_no)+"/fi"+str(i) for i in range(4)]
    cm, thr_prre, pre, rec, thr_roc, fpr, tpr = aT.evaluate_model_for_folders(test_folders, model_name+str(fold_no), "svm", "fi1")
```

```python
for i in range(3):
    train_fold(i)
    test_fold(i)
```
